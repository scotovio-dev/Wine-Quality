```python
import pandas as pd
```

#### Importing Datasets

```python
red = pd.read_csv("winequality-red.csv", sep=";")
white = pd.read_csv("winequality-white.csv", sep=";")
```

### EDA

```python
print("Red")
display(red.head())
print("White")
display(white.head())
```

```python
print("Red shape: ", red.shape)
print("White shape: ", white.shape)
```

#### Nulls

```python
missing_df = pd.DataFrame({
    'red': red.isna().sum(),
    'white': white.isna().sum()
})

missing_df = missing_df.style.set_caption('Missing Values Summary')
display(missing_df)
```

#### Duplicates

```python
red_dupl = red.duplicated().sum()
white_dupl = white.duplicated().sum()
print("red: ",red_dupl)
print("white: ", white_dupl)
```

```python
red.drop_duplicates(inplace=True)
white.drop_duplicates(inplace=True)

print("Red shape: ", red.shape)
print("White shape: ", white.shape)
```
